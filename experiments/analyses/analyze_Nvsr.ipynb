```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
inferences = ['torch','EM']

for inference in inferences:
    df = pd.read_csv('../../data/results/torchvsEM_116_results_N_increasing2/Real116_'+inference+'_Watson.csv')
    plt.figure(figsize=(6,3))
    sns.boxplot(x='N',y='test_NMI',hue='K',data=df)
    plt.title('Real data (p=116), stability over initializations: '+inference+' Watson')
```

```python
inferences = ['torch','EM']
fig,axs = plt.subplots(7,2,figsize=(8,16),layout='constrained')
for i,inference in enumerate(inferences):
    df = pd.read_csv('../../data/results/torchvsEM_synthetic_results_N_increasing/Synthetic_'+inference+'_Watson.csv')
    df['N'] = np.log(df['N'])
    for pp,p in enumerate([3,10,25,50,100,500,1000]):
        sns.lineplot(x='N',y='test_NMI',hue='K',data=df[df['p']==p],ax=axs[pp,i],linewidth=3,markers=True,style='K',dashes=False)
        axs[pp,i].set_title('Synthetic data '+inference+' Watson p='+str(p))
        axs[pp,i].set_ylim(0,1.05)
        axs[pp,i].set_xticks(np.log([50,100,500,1000,5000,10000]),[50,100,500,1000,5000,10000],rotation=15)
```

```python
inferences = ['torch','EM']
modelnames = ['ACG','MACG']
Ks = [2,5,10]
for m,modelname in enumerate(modelnames):
    fig,axs=plt.subplots(2,3,layout='constrained',figsize=(9,6))
    for i,inference in enumerate(inferences):
        df = pd.read_csv('../../data/results/torchvsEM_116_results_N_increasing2/Real116_'+inference+'_'+modelname+'.csv')
        for k,K in enumerate(Ks):
            df2 = df[df['K']==K]
            #take average over test_NMI
            df2 = df2.groupby(['modelname','init_method','ACG_rank','N']).mean().reset_index()
            # make ACG_rank categorical and specify category order
            df2 = df2.pivot(index='ACG_rank',columns='N',values='test_NMI')
            sns.heatmap(df2,cmap='viridis',cbar_kws={'label': 'test_NMI'},vmin=0,vmax=1,ax=axs[i,k]);
            # sns.boxplot(x='N',y='test_NMI',hue='K',data=df)
            axs[i,k].set_title('Clustering stability, '+inference+' '+modelname+' K='+str(K));
```
