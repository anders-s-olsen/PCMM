```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
inferences = ['torch','EM']

for inference in inferences:
    df = pd.read_csv('../../data/results/torchvsEM_116_results_N_increasing/Real116_'+inference+'_Watson.csv')
    plt.figure(figsize=(6,3))
    sns.boxplot(x='N',y='test_NMI',hue='K',data=df)
    plt.title('Real data Stability over initializations'+inference+' Watson')
```

```python
inferences = ['torch','EM']

for inference in inferences:
    df = pd.read_csv('../../data/results/torchvsEM_synthetic_results_N_increasing/Synthetic_'+inference+'_Watson.csv')
    plt.figure(figsize=(6,3))
    sns.boxplot(x='N',y='test_NMI',hue='K',data=df)
    plt.ylim(0.4,0.8)
    plt.title('Synthetic data '+inference+' Watson')
```

```python
inferences = ['torch','EM']
modelnames = ['ACG','MACG']
Ks = [2,5,10]
for modelname in modelnames:
    for inference in inferences:
        df = pd.read_csv('../../data/results/torchvsEM_116_results_N_increasing/Real116_'+inference+'_'+modelname+'.csv')
        for K in Ks:
            plt.figure(figsize=(3,3));
            df2 = df[df['K']==K]
            #take average over test_NMI
            df2 = df2.groupby(['modelname','init_method','ACG_rank','N']).mean().reset_index()
            # make ACG_rank categorical and specify category order
            df2 = df2.pivot(index='ACG_rank',columns='N',values='test_NMI')
            sns.heatmap(df2,cmap='viridis',cbar_kws={'label': 'test_NMI'},vmin=0,vmax=1);
            # sns.boxplot(x='N',y='test_NMI',hue='K',data=df)
            plt.title('Stability over initializations, '+inference+' '+modelname+' K='+str(K));
```
