```python
def get_sphere(resolution=40):
    u, v = np.mgrid[0:2*np.pi:resolution*1j, 0:np.pi:resolution/2*1j]
    x = np.cos(u)*np.sin(v)
    y = np.sin(u)*np.sin(v)
    z = np.cos(v)
    return x, y, z

def draw_sphere(ax):
    # draw sphere
    x,y,z = get_sphere(resolution=40)
    ax.plot_wireframe(x, y, z, color="k", alpha=0.1)
    # remove ticks
    ax.set_xticks([-1,1])
    ax.set_yticks([-1,1])
    ax.set_zticks([-1,1])
    ax.set_aspect('equal')
    ax.set_xlabel(r'$V_{\cdot,1}$',labelpad=0.01)
    ax.set_ylabel(r'$V_{\cdot,2}$',labelpad=0.01)
    ax.set_zlabel(r'$V_{\cdot,3}$',labelpad=0.01)

    ax.tick_params(axis="y", pad=0.01)
    ax.tick_params(axis="z", pad=0.01)
    ax.tick_params(axis="x", pad=0.01)

    # keep the three main axes but remove the gray area in the middle of the 3d plot
    ax.xaxis.pane.fill = False
    ax.yaxis.pane.fill = False
    ax.zaxis.pane.fill = False
    ax.grid(False)
    return ax

def add_sphere_points_color(ax, P,color='b'):

    ax.scatter(P[0], P[1], P[2], color=color, s=1, alpha=0.5)

    if P.shape[1]<3:
        ax.plot([0, P[0,0]], [0, P[1,0]], [0, P[2,0]], color=color)
    
    return ax

def add_sphere_centroids_color(ax, centroids, color):
    ax.scatter(centroids[0], centroids[1], centroids[2], color=color, s=100,marker='x')
    return ax
```

```python
from src.DMM_EM.WatsonEM import Watson
fig,axs = plt.subplots(1,2,figsize=(10,5),subplot_kw={'projection': '3d'})

# in df, find one of the rows corresponding to kmeans and extract the labels
labels = np.int32(df[df['method']=='Watson'].iloc[0]['label'])
centroids = df[df['method']=='Watson'].iloc[0]['centroids']

Watson = Watson(K=2,p=3,params=centroids)

x,y,z = get_sphere(resolution=100)
X = np.array([x.flatten(),y.flatten(),z.flatten()]).T

log_pdf = Watson.log_pdf(X)
log_pdf = log_pdf.reshape((2,x.shape[0],x.shape[1]))
norm_log_pdf = np.zeros_like(log_pdf)
norm_log_pdf[0] = (log_pdf[0]-log_pdf[0].min())/(log_pdf[0].max()-log_pdf[0].min())
norm_log_pdf[1] = (log_pdf[1]-log_pdf[1].min())/(log_pdf[1].max()-log_pdf[1].min())

norm_log_pdf[0,norm_log_pdf[0]<0.75] = 0
norm_log_pdf[1,norm_log_pdf[1]<0.75] = 0

# show log_pdf as contour on sphere in 3d
axs[0].plot_surface(x,y,z, 
                    facecolors=plt.cm.Blues(norm_log_pdf[0]), alpha=0.5, antialiased=False)
axs[0] = draw_sphere(axs[0])
axs[0].plot_surface(x,y,z, 
                    facecolors=plt.cm.Reds(norm_log_pdf[1]), alpha=0.5, antialiased=False)
axs[0] = draw_sphere(axs[0])
# don't show contours where the value is below zero
# axs[0].plot_surface(x,y,z, facecolors=log_pdf[0], levels=0, colors='k', alpha=0.5)

# axs[0] = draw_sphere(axs[0])
# axs[0] = add_sphere_points_color(axs[0], data[labels[0]==1,:,0].T,color='k')
# axs[0] = add_sphere_points_color(axs[0], data[labels[1]==1,:,0].T,color='gray')
# axs[0] = add_sphere_centroids_color(axs[0], centroids[0], color='k')
# axs[0] = add_sphere_centroids_color(axs[0], centroids[1], color='gray')

# axs[1] = draw_sphere(axs[1])
# axs[1] = add_sphere_points_color(axs[1], data[labels[0]==1,:,0].T,color='k')
# axs[1] = add_sphere_points_color(axs[1], data[labels[1]==1,:,0].T,color='gray')
# axs[1] = add_sphere_centroids_color(axs[1], centroids[0], color='k')
# axs[1] = add_sphere_centroids_color(axs[1], centroids[1], color='gray')
# axs[1].view_init(30,30)
```
